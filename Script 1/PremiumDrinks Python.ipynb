DATOS ESTADISTICOS BASICOS:

```python
def calcular_estadisticas(data): 
    for columna in data.columns:
        if data[columna].dtype == 'float64' or data[columna].dtype == 'int64':
            print(columna)
            print('Mediana:', data[columna].median())
            print('Varianza: ', np.around((data[columna].var()),decimals=2)) # Casteo (redondeo) a 2 decimales
            print('Rango:', data[columna].max() - data[columna].min())
            print('Moda:', data[columna].mode()[0])
            print('--------------------------------')

calcular_estadisticas(df)
```

```python


# Leer el archivo CSV
DataInventFinal = pd.read_csv(r"C:\Users\usuario\Desktop\PFinal\EndInvFINAL12312016_Clean.csv")


# Agrupar por ciudad y sumar las ventas usando la columna 'onHand'
ventas_por_ciudad = DataInventFinal.groupby('City')['onHand'].sum().reset_index()

# Ordenar por ventas de mayor a menor y seleccionar las ciudades top
top_ciudades = ventas_por_ciudad.sort_values(by='onHand', ascending=False).head(10)

# Crear el gráfico de barras
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top_ciudades['City'], top_ciudades['onHand'], color='blue')

# Agregar etiquetas y título
plt.xlabel('Ciudad')
plt.ylabel('Ventas (onHand)')
plt.title('Top 10 de ventas por ciudad')
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python



# Agrupar por ciudad y descripción del producto, luego sumar las cantidades
productos_por_ciudad = DataInventFinal.groupby(['City', 'Description'])['onHand'].sum().reset_index()

# Obtener el top 10 de productos por ciudad
top_productos_por_ciudad = productos_por_ciudad.loc[productos_por_ciudad.groupby('City')['onHand'].idxmax()]

# Mostrar el resultado
print(top_productos_por_ciudad)

# Crear gráfico de barras para visualizar el top 10 de productos por ciudad
plt.figure(figsize=(12, 6))
plt.bar(top_productos_por_ciudad['City'] + ' - ' + top_productos_por_ciudad['Description'], 
        top_productos_por_ciudad['onHand'], color='green')

# Agregar etiquetas y título
plt.xlabel('Ciudad - Producto')
plt.ylabel('Cantidad Vendida (onHand)')
plt.title('Top 10 Productos Más Vendidos por Ciudad')
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python


# Agrupar por ciudad y descripción del producto, luego sumar las cantidades
productos_por_ciudad = DataInventFinal.groupby(['City', 'Description'])['onHand'].sum().reset_index()

# Obtener el top 3 de productos por ciudad
top_productos_por_ciudad = productos_por_ciudad.sort_values(by='onHand', ascending=False)

# Seleccionar los 3 principales por ciudad
top_3_productos_por_ciudad = top_productos_por_ciudad.groupby('City').head(3)

# Mostrar el resultado en una tabla
print(top_3_productos_por_ciudad)
```

Top 3 ciudades que venden mas y 3 que venden menos

```python


# Agrupar por ciudad y sumar las cantidades vendidas
ventas_por_ciudad = DataInventFinal.groupby('City')['onHand'].sum().reset_index()

# Ordenar las ciudades por ventas
ventas_por_ciudad = ventas_por_ciudad.sort_values(by='onHand', ascending=False)

# Obtener el top 3 y el bottom 3 de ciudades
top_3_ciudades = ventas_por_ciudad.head(3)
bottom_3_ciudades = ventas_por_ciudad.tail(3)

# Combinar top y bottom en un solo DataFrame
top_bottom_ciudades = pd.concat([top_3_ciudades, bottom_3_ciudades])

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(top_bottom_ciudades['City'], top_bottom_ciudades['onHand'], color=['green']*3 + ['red']*3)

# Agregar etiquetas y título
plt.xlabel('Ciudad')
plt.ylabel('Cantidad Vendida (onHand)')
plt.title('Top 3 y Bottom 3 Ciudades por Ventas')
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
